```python
def generate_random_action():
    return np.random.uniform(0, 2, size=(1, 2))
```

```python
def get_action_singleagent()-> ActionTuple:
    if np.random.random() < EPS : tuple.add_discrete(generate_random_action())
    else : tuple.add_discrete(np.ones((1,2)))
    return tuple
```
