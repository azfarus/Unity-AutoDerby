```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
group_rewards_PPO: np.ndarray
elo_PPO: np.ndarray

group_rewards_A2C: np.ndarray
elo_A2C: np.ndarray
```

```python
# filename = 'A2C_output.txt'
filename = 'PPO_output.txt'
training_data_list=[]
with open(filename , 'r') as train_data:
    for line in train_data:
        data_list=[]
        for word in line.split(' '):
            try:
                data_list.append(float(word))
            except:
                pass
        if len(data_list) > 0 : training_data_list.append(data_list)

result=np.array(training_data_list).T
print(result)

group_rewards_PPO = result[1]
elo_PPO = result[0]
```

```python
filename = 'A2C_output.txt'
# filename = 'PPO_output.txt'
training_data_list=[]
with open(filename , 'r') as train_data:
    for line in train_data:
        data_list=[]
        for word in line.split(' '):
            try:
                data_list.append(float(word))
            except:
                pass
        if len(data_list) > 0 : training_data_list.append(data_list)

result=np.array(training_data_list).T
print(result)

group_rewards_A2C = result[2]
elo_A2C = result[1]
```

```python
plt.plot(range(len(elo_A2C)) , elo_A2C )
plt.xlabel('Epochs')
plt.ylabel('ELO')
plt.title('A2C ELO')
plt.savefig('a2c_graphs/elo')
plt.show()
plt.plot(range(len(elo_PPO)) , elo_PPO )
plt.xlabel('Epochs')
plt.ylabel('ELO')
plt.title('PPO ELO')
plt.savefig('ppo_graphs/elo')
plt.show()
```

```python
plt.plot(range(len(group_rewards_A2C)) , group_rewards_A2C )
plt.xlabel('Epochs')
plt.ylabel('Group Rewards')
plt.title('A2C Group Rewards')
plt.savefig('a2c_graphs/grouprewards')
plt.show()
plt.plot(range(len(group_rewards_PPO)) , group_rewards_PPO )
plt.xlabel('Epochs')
plt.ylabel('Group Rewards')
plt.title('PPO Group Rewards')
plt.savefig('ppo_graphs/grouprewards')
plt.show()
```
